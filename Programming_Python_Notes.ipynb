###Matrix Mult

````python
#Matrix Multiplication

r1=[1,2,3]
r2=[4,5,6]
r3=[7,8,9]

s1=[1,2,1]
s2=[6,2,3]
s3=[4,2,1]

A=[]
B=[]
A.append(r1)
A.append(r2)
A.append(r3)

B.append(s1)
B.append(s2)
B.append(s3)

C=[[0,0,0],[0,0,0],[0,0,0]]
dim=3

for i in range(dim):
  for j in range(dim):
    for k in range(dim):
      C[i][j]=C[i][j]+A[i][k]*B[k][j]
print(C)
````

````python
import numpy
X=numpy.asmatrix(A)
Y=numpy.asmatrix(B)
print(X*Y)
````

````python
#Matrix Multiplication using functions

def initialize_mat(dim):
  C=[]
  for i in range(dim):
    C.append([])
  for i in range(dim):
    for j in range(dim):
      C[i].append(0)
  return C

initialize_mat(3)
````

````python
def dot_product(u,v):
  dim=len(u)
  ans=0
  for i in range(dim):
    ans=ans+(u[i]*v[i])
  return ans

l1=[1,2,3]
l2=[3,5,6]
print(dot_product(l1,l2))
````

````python
def row(M,i):
  dim=len(M)
  l=[]
  for k in range(dim):
    l.append(M[i][k])
  return l

def column(M,j):
  dim=len(M)
  l=[]
  for k in range(dim):
    l.append(M[k][j])
  return l
````

````python
def mat_mul(A,B):
  dim=len(A)
  C=initialize_mat(dim)
  for i in range(dim):
    for j in range(dim):
      C[i][j]=dot_product(row(A,i),column(B,j))

  return C
````

````python
#Problem 1

def upper(s):
  upper=0
  for c in s:
    if(c.isupper()):
      upper+=1
  return(upper)

def lower(s):
  lower=0
  for c in s:
    if(c.islower()):
      lower+=1
  return lower

def character(s):
  chars=0
  for c in s:
      chars+=1
  return chars

def words(s):
  words=1
  for c in s:
    if(c==' '):
      words+=1
  return words

sentence=input("Enter the sentence")
uLetters=upper(sentence)
print(f'\nTotal number of upper case characters: {uLetters}')
lLetters=lower(sentence)
print(f'\nTotal number of lower case characters: {lLetters}')
chars=character(sentence)
print(f'\nTotal number of characters: {chars}')
words=words(sentence)
print(f'\nTotal number of wordss: {words}')
````

````python
#Problem 3

#Checking a triangle
#Approach 1: Using the distance formula: The sum of two smaller lengths is greater than the max length

#Approach 2: Using slope method

import math

def slope(xi,yi,xj,yj):
  if(xi==xj):
    return math.inf
  else:
    return (yj-yi)/(xj-xi)

x1=float(input("Enter x coordinate"))
y1=float(input("Enter y coordinate"))
x2=float(input("Enter x coordinate"))
y2=float(input("Enter y coordinate"))
x3=float(input("Enter x coordinate"))
y3=float(input("Enter y coordinate"))
s1=slope(x1,y1,x2,y2)
s2=slope(x2,y2,x3,y3)
if(s1==s2):
  print("Not a triangle")
else:
  print("Triangle")

````

````python
''' This program considers an input file and encrypts it by using caesar cipher. By that we mean, we shift the letters by 3 units. For example a becomes d
b becomes e and so on'''
import string

def create_caesar_dictionary():
  l=string.ascii_lowercase
  l=list(l)
  d={}
  for i in range(len(l)):
    d[l[i]]=l[(i+3)%26]
  return d

f=open("sherlock.txt",'r')
g=open("cipher.txt",'w')
d=create_caesar_dictionary()

c=f.read(1)
while(c!=''):
  g.write(d[c])
  c=f.read(1)

f.close()
g.close()
````

````python
#checks if a list has 0 in it using recursion
def check0(L):
  if(len(L)==0):
    return 0
  if(L[0]==0):
    return 1
  else:
    return check0(L[1:len(L)])

print(check0([1,2,3,11,14,15]))

````

````python
#Sorting using Recursion

def mini(L):# find the minimum element in the list L
  mini=L[0]
  for x in L:
    if(x<mini):
      mini=x
  return mini

def Sort(L): # recursively sort the list L
  if (L==[] or len(L)==1):
    return L# if the list is empty or there is nothing to sort
  m=mini(L)# contains the minimum most element
  L.remove(m)#we remove that element from L
  return [m]+Sort(L) #recursively sorting the smaller list

L=[5,7,6,3,5,69,76]
print(Sort(L))
````

````python
def binary_search(L,k):
  '''This function is an alt for obvious search it does similar but in an efficient way
  Saves time. Binary search'''
  begin=0 #first element in L. L[0]
  end=len(L)-1 #Last element in L
  while(end-begin>1):
    mid=(begin+end)//2
    if(L[mid]==k):
      return 1
    if(L[mid]<k):
      begin=mid+1
    if(L[mid]>k):
      end=mid-1
  if(L[begin]==k or L[end]==k):
    return 1
  else:
    return 0

L=[12,15,100,121,1001,1024,2016,2021]

print(binary_search(L,1001))
````

````python
#Binary Search using Recursion
def rbinary_search(L,k,begin,end):
    if(begin==end): # if begin and end a re same then we need to just check L[begin]
      if(L[begin]==k):
        return 1
      else:
        return 0
    if(end-begin==1): # if they are consecutive then check them individually
      if(L[begin]==k or L[end]==k):
        return 1
      else:
        return 0
    if(end-begin>1):
      mid=(end+begin)//2
      if(L[mid]>k):
        end=mid-1
      if(L[mid]<k):
        begin=mid+1
      if(L[mid]==k):
        return 1
    if(end-begin<0):
        return 0
    return rbinary_search(L,k,begin,end)

L=list(range(1000))
rbinary_search(L,255,0,999)
````

````python
import pandas as pd
scores = pd.read_csv('scores.csv')
print(scores,type(scores))# the answer to this will be - Data Frame

print(scores['Name'],type(scores['Name']))# - Series: Any specific column from the dataframe

print(scores.head())#will print top 5 rows

print(scores.tail())# will print last 5 rows

print(scores[scores['Name']=='Nisha']) # to extract a particular row value
#accessed the data frame and in that data frame it accesses Name==Nisha

#Maxmimum total based on Gender
print(scores[scores['Gender']=='M']['Total'].max())
print(scores[scores['Gender']=='F']['Total'].max())

#Students grading
print(scores[scores['Physics']>85].shape[0])
print(scores[scores['Physics'].between(70,85)].shape[0])
print(scores[scores['Physics'].between(60,70)].shape[0])
print(scores[scores['Physics']<60].shape[0])

#Further Conditioning
print(scores[(scores['Gender']=='M') & (scores['Physics']>85)].shape[0])
print(scores[(scores['Gender']=='F') & (scores['Physics']>85)].shape[0])

#Further Complications
subject=['Maths','Physics','Chem']
for sub in subject:
  print('Above 85 in', sub)
  print(scores[(scores['Gender']=='M') & (scores['sub']>85)].shape[0])
  print(scores[(scores['Gender']=='F') & (scores['sub']>85)].shape[0])

#For finding out above average students
subject=['Maths','Physics','Chem']
for sub in subject:
  print('Above avg in', sub)
  avg=scores[sub].mean()
  print(scores[(scores['Gender']=='M') & (scores['sub']>avg)].shape[0])
  print(scores[(scores['Gender']=='F') & (scores['sub']>avg)].shape[0])

#GroupBy

print(scores.groupby('Gender').groups)

#For finding out above average students using GroupBy
subject=['Maths','Physics','Chem']
for sub in subject:
  print('Above avg in', sub)
  avg=scores[sub].mean()
  print(scores[scores['sub']>avg].groupby('Gender').groups)

````
